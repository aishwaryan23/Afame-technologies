## DROPPING COLUMNS WHICH ARE UNNECESSARY

```python
new_data=data.drop(data.columns[[6,8,10,13,16,18,21,25,27,30,34]],axis=1)
new_data
```

## RENAMING THE COLUMNS

```python
new_data.rename(columns={'EmployeeNumber': 'EmpID'}, inplace=True)
new_data.rename(columns={'Department':'Dept'},inplace=True)
new_data
```

### FINDING THE UNIQUE VALUES 

```python
new_data['Gender'].unique()
```

```python
new_data['Gender'] = new_data['Gender'].map({'Female':'F','Male':'M'})
```

```python
new_data['Dept'].unique()
```

```python
new_data['Dept']=new_data['Dept'].map({'Research & Development':'R&D','Sales':'Sales','Human Resources':'HR'})
```

```python
new_data
```

```python
new_data.describe(include="O")
```

## GRAPH OF ATTRITION

```python
sns.countplot(x=new_data.Attrition)
plt.show()
```

## FACTORS AFFECTING ATTRITION

```python
sns.countplot(hue=new_data.Attrition,x=new_data.BusinessTravel)
plt.show()
```

```python
sns.countplot(hue=new_data.Attrition,x=new_data.Dept)
plt.show()
```

```python
sns.countplot(x=new_data.Attrition,hue=new_data.EducationField)
plt.show()
```

```python
sns.countplot(x=new_data.Attrition,hue=new_data.Gender)
plt.show()
```

```python
sns.countplot(hue=new_data.Attrition,x=new_data.OverTime)
plt.show()
```

```python
plt.figure(figsize=(20,10),facecolor='white')
sns.countplot(x='JobRole',hue='Attrition',data=new_data)
plt.xlabel('JobRole',fontsize=20)
```
